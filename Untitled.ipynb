#Handling Outliers

```python
Q1= data.quantile(0.25)
Q3= data.quantile(0.75)
IER = Q3 - Q1
print(IER)
```

```python
##Removing Outliers
data = data[~((data<(Q1-1.5 * IER))| (data>(Q3 + 1.5 * IER))).any(axis=1)]
```
